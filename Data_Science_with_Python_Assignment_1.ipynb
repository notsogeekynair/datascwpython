```python
# -*- coding: utf-8 -*-
"""
Created on Mon Nov  5 14:37:29 2018

@author: epinsky
"""
# run this !pip install pandas_datareader
from pandas_datareader import data as web
import os
import pandas as pd
import yfinance as yf
from google.colab import files



def get_stock(ticker, start_date, end_date, s_window, l_window):
    try:
        df = yf.download(ticker, start=start_date, end=end_date)
        df['Return'] = df['Adj Close'].pct_change()
        df['Return'].fillna(0, inplace=True)
        df['Date'] = df.index
        df['Date'] = pd.to_datetime(df['Date'])
        df['Month'] = df['Date'].dt.month
        df['Year'] = df['Date'].dt.year
        df['Day'] = df['Date'].dt.day
        for col in ['Open', 'High', 'Low', 'Close', 'Adj Close']:
            df[col] = df[col].round(2)
        # df['Weekday'] = df['Date'].dt.weekday_name
        df['Weekday'] = df['Date'].dt.day_name()
        df['Week_Number'] = df['Date'].dt.strftime('%U')
        df['Year_Week'] = df['Date'].dt.strftime('%Y-%U')
        df['Short_MA'] = df['Adj Close'].rolling(
            window=s_window, min_periods=1).mean()
        df['Long_MA'] = df['Adj Close'].rolling(
            window=l_window, min_periods=1).mean()
        col_list = ['Date', 'Year', 'Month', 'Day', 'Weekday',
                    'Week_Number', 'Year_Week', 'Open',
                    'High', 'Low', 'Close', 'Volume', 'Adj Close',
                    'Return', 'Short_MA', 'Long_MA']
        num_lines = len(df)
        df = df[col_list]
        print('read ', num_lines, ' lines of data for ticker: ', ticker)
        return df
    except Exception as error:
        print(error)
        return None

os.chdir('/content/drive/MyDrive/BU Fall 24/Data Science with Python/Assignments/week_1_homework')
here = os.getcwd()
input_dir = os.path.abspath(os.path.join(here, os.pardir))
tickers = ['SPY','NVDA']
for ticker in tickers:
    try:
        output_file = os.path.join(input_dir, ticker + '.csv')
        df = get_stock(ticker, start_date='2016-01-01', end_date='2020-12-31',
                       s_window=14, l_window=50)
        df.to_csv(output_file, index=False)
        print('wrote ' + str(len(df)) + ' lines to file: ' + output_file)
    except Exception as e:
        print(e)
        print('failed to get Yahoo stock data for ticker: ', ticker)
```

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Nov  5 14:37:29 2018

@author: epinsky
this scripts reads your ticker file (e.g. MSFT.csv) and
constructs a list of lines
"""

"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #1
Description :for each of the 5 years, compute the mean and standard
"""
import os
from google.colab import files
ticker='NVDA'
input_dir = r'/content/drive/MyDrive/BU Fall 24/Data Science with Python/Assignments'
ticker_file = os.path.join(input_dir, ticker + '.csv')

try:
    with open(ticker_file) as f:
        lines = f.read().splitlines()
    print('opened file for ticker: ', ticker)
    """    your code for assignment 1 goes here
    """
    title = lines[0].split(',')

    data =[]
    for line in lines[1:]:
        columns = line.split(',')

        date=columns[0]
        year=int(columns[1])
        month=int(columns[2])
        day=int(columns[3])
        weekday=columns[4]
        returns = float(columns[13])

        data.append({'date': date,
                     'year': year,
                     'month': month,
                     'day': day,
                     'weekday': weekday,
                     'returns' : returns
                     })
    #print(data[0:7])

    years = {2016,2017,2018,2019,2020}
    weekdays = {'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'}

    results ={}

    for i in data:
        year = i['year']
        weekday = i['weekday']
        returns = i['returns']

        if year in years and weekday in weekdays:
          key =(year,weekday)
          if key not in results:
            results[key]={'R':[],'R-':[], 'R+':[]}
          results[key]['R'].append(returns)
          if returns < 0:
            results[key]['R-'].append(returns)
          else:
            results[key]['R+'].append(returns)
    def mean(values):
      return sum(values)/len(values) if values else 0

    def std_dev(values,mean):
      return (sum((x - mean) ** 2 for x in values) / len(values)) ** 0.5 if values else 0

    for (year, weekday), sets in results.items():
      print(f"Year: {year}, Weekday: {weekday}")
      for set_name in ['R', 'R-', 'R+']:
          returns = sets[set_name]
          mean_value = mean(returns)
          std_dev_value = std_dev(returns, mean_value)
          print(f"  {set_name}: Mean = {mean_value:.4f}, Std Dev = {std_dev_value:.4f}")

except Exception as e:
    print(e)
    print('failed to read stock data for ticker: ', ticker)
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #1
Description :creating table for both stocks
"""
from google.colab import files
def read_csv_file(file_path):
    try:
        with open(file_path, 'r') as f:
            lines = f.readlines()
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None

    data = []
    for line in lines[1:]:
        columns = line.split(',')
        data.append({
            'date': columns[0],
            'year': int(columns[1]),
            'month': int(columns[2]),
            'day': int(columns[3]),
            'weekday': columns[4],
            'return': float(columns[13])
        })
    return data

uploaded = files.upload()
nvda = read_csv_file('NVDA.csv')
spy = read_csv_file('SPY.csv')

#print(nvda[0:7])
print(spy[0:7])
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #3
Description :Compute the aggregate table across all 5 years,
"""
def agg(data):
  aggregate = {
      'Monday':[],
      'Tuesday':[],
      'Wednesday':[],
      'Thursday':[],
      'Friday':[]
  }
  for i in data:
    weekday = i['weekday']
    returns = i['return']
    aggregate[weekday].append(returns)
  return aggregate


nvda_agg = agg(nvda)
spy_agg = agg(spy)
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #3
Description :Compute the aggregate table across all 5 years
"""
def tabledata(agregated_data):
  results = {}
  for weekday, returns in agregated_data.items():
    rn = [r for r in returns if r<0]
    rp = [r for r in returns if r>=0]

    meanr = mean(returns)
    stdr = std_dev(returns,meanr)


    meanrn = mean(rn)
    stdrn = std_dev(rn,meanrn)

    meanrp = mean(rp)
    stdrp = std_dev(rp,meanrp)

    results[weekday] = {
        'Mean R': meanr,
        'Std Dev R': stdr,
        'Mean R-': meanrn,
        'Std Dev R-': stdrn,
        'R- Days': len(rn),
        'Mean R+': meanrp,
        'Std Dev R+': stdrp,
        'R+ Days': len(rp),
    }
  return results

nvda_table = tabledata(nvda_agg)
spy_table = tabledata(spy_agg)
print(nvda_table)
print(spy_table)
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #4
Description : You listen to the oracle and follow its advice.
"""
#q4
def oracle(data,p):
  for i in data :
    stock_return = i['return']

    if stock_return >0:
      p = p + (p*stock_return)
  return p

nvda_final =oracle(nvda,100)
spy_final =oracle(spy,100)

print('Return for NVDA on last day : ',nvda_final)
print('Return for SPY on last day : ',spy_final)
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #5
Description : Consider "buy-and-hold" strategy:
"""
#q5

def disobey_oracle(data, p):
  for i in data:
    stock_return =i['return']
    p = p + (p*stock_return)
  return p

nvda_disobey = disobey_oracle(nvda,100)
spy_disobey = disobey_oracle(spy,100)

print('Buy and hold strategy for NVDA : ', nvda_disobey)
print('Buy and hold strategy for SPY : ',spy_disobey)
```

```python
"""
Varun Nair
Class: CS 677
Date: 09/13/2024
Homework Problem #6
Description : Your oracle got very upset that you did not follow its advice.
"""
#q6
##qa
#Missing the 10 best trading days
nvdasorted = sorted(nvda, key=lambda x: x['return'])
spysorted = sorted(spy, key=lambda x: x['return'])

qa = nvdasorted[0:len(nvdasorted)-10]
qaspy = spysorted[0:len(spysorted)-10]

nvda_miss = disobey_oracle(qa,100)
spy_miss = disobey_oracle(qaspy,100)

print('Investing in NVDA while missing the best 10 days : ',nvda_miss)
print('Investing in SPY while missing the best 10 days : ',spy_miss)

##qb
#Assuming we are ONLY investing in the 10 worst days
qb = nvdasorted[0:10]
qbspy = spysorted[0:10]

nvda_worst = disobey_oracle(qb,100)
spy_worst = disobey_oracle(qbspy,100)

print('Investing in NVDA only on the worst 10 days : ',nvda_worst)
print('Investing in SPY only on the worst 10 days : ' ,spy_worst)

##qc
#Investing in 5 worst days and missing the 5 best days
nvda_qc = nvdasorted[5:len(nvdasorted)-5]
spy_qc = spysorted[5:len(spysorted)-5]

nvda_last = disobey_oracle(nvda_qc,100)
spy_last = disobey_oracle(spy_qc,100)

print('Investing in NVDA on the 5 worst days and missing the 5 best days : ',nvda_last)
print('Investing in SPY on the 5 worst days and missing the 5 best days : ',spy_last)
```
